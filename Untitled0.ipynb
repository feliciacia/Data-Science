###BIRCH

```python
from sklearn.cluster import Birch
```

```python
from sklearn import metrics
K=range(3,16)
S=[]
for k in K:
    # Set the model and its parameters
    model = Birch(n_clusters=k, threshold= 0.01)
    # Fit the model 
    labels = model.fit_predict(normalized_data)
    # Calculate Silhoutte Score and append to a list
    S.append(metrics.silhouette_score(normalized_data, labels, metric='euclidean'))
plt.figure(figsize=(16,8), dpi=300)
plt.plot(K, S, 'bo-', color='black')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('Identify the number of clusters using Silhouette Score')
plt.show()
```

```python
birch = Birch(threshold=0.01,n_clusters=8)
```

```python
group_birch = birch.fit_predict(data)
```

```python
plt.scatter(x_axis,y_axis,c = group_birch)
plt.show()
```

```python
group_birch_pd = pd.DataFrame(group_birch)
group_birch_pd.head()
print(group_birch_pd.value_counts())
```

```python
answer_birch = []
count1 = 0
for i in range(len(test)):
  if(group_birch[test["0"][i]]==group_birch[test["1"][i]]):
    answer_birch.append(1)
    count1 = count1+1
  else:
    answer_birch.append(0)
print(answer_birch)
print(count1)
```

###K-MEANS

```python
from sklearn.cluster import KMeans
```

```python
from sklearn import metrics
K=range(3,16)
S=[]
for k in K:
    # Set the model and its parameters
    model = KMeans(n_clusters=k, random_state=31)
    # Fit the model 
    labels = model.fit_predict(data_scaled_pd)
    # Calculate Silhoutte Score and append to a list
    S.append(metrics.silhouette_score(data_scaled_pd, labels, metric='euclidean'))
plt.figure(figsize=(16,8), dpi=300)
plt.plot(K, S, 'bo-', color='black')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.title('Identify the number of clusters using Silhouette Score')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 15, random_state=31,init='k-means++',max_iter=300,verbose=0)
```

```python
plt.scatter(x_axis,y_axis,c = group_kmeans)
plt.show()
```

```python
group_kmeans_pd = pd.DataFrame(group_kmeans)
group_kmeans_pd.head()
print(group_kmeans_pd.value_counts())
```

```python
answer_kmeans = []
count1 = 0
for i in range(len(test)):
  if(group_kmeans[test["0"][i]]==group_kmeans[test["1"][i]]):
    answer_kmeans.append(1)
    count1 = count1+1
  else:
    answer_kmeans.append(0)
print(answer_kmeans)
print(count1)
```

###Voting

```python
final_ans = np.zeros(400)
for i in range(2000):
  randomNum = np.random.randint(0,100000)
  kmeans = KMeans(n_clusters = int(randomNum%3)*2 + 11, random_state=randomNum,init='k-means++',max_iter=300,verbose=0)
  group_kmeans = kmeans.fit_predict(data_scaled_pd)
  answer_kmeans = []
  count1 = 0
  for i in range(len(test)):
    if(group_kmeans[test["0"][i]]==group_kmeans[test["1"][i]]):
      answer_kmeans.append(1)
      count1 = count1+1
    else:
      answer_kmeans.append(0)
  final_ans = final_ans + answer_kmeans
  print(count1)
print(final_ans)
```

```python
final_ans_divided = final_ans / 2000.0
ans = []
count1 = 0
for i in range(len(final_ans_divided)):
  if final_ans_divided[i]>0.5:
    ans.append(1)
    count1 = count1 + 1
  else:
    ans.append(0)
print(ans)
print(count1)
```
